```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
# LinearRegression 普通最小二乘线性回归
import sklearn.linear_model
# 导入波士顿房价数据集
from sklearn.datasets import load_boston
# 实验中用到的回归模型
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.utils import check_random_state
# 数据划分
from sklearn.model_selection import train_test_split
# 均方误差
from sklearn.metrics import mean_squared_error
plt.rc('font', size=14)
import warnings
warnings.filterwarnings("ignore")

# 读取波士顿房价数据集
boston = load_boston() #加载数据
boston_dataset = pd.DataFrame(boston.data, columns=boston.feature_names)
# MEDV是房价中位数即标签
boston_dataset['MEDV'] =  boston.target
# 展示前五条数
boston_dataset.head(5)
print(boston_dataset.head(5))

# 划分数据集
x_train, x_test, y_train, y_test = train_test_split(boston.data, boston.target,random_state=1)


# LinearRegression
print('--------------------LinearRegression--------------------')
model2 = LinearRegression()
model_f = model2.fit(x_train, y_train)
# 训练后模型截距
print('训练后模型截距')
print(model2.intercept_)

# 训练后模型权重
print('训练后模型权重')
print(model2.coef_) 

m = model2.coef_
b = model2.intercept_
# 根据得到的系数和截距打印回归方程print(' y = {0} * x + {1}'.format(m, b))
print(r'根据得到的系数和截距打印回归方程')
print(' y = {0} * x + {1}'.format(m, b))

# 使用模型预测
plt.scatter(y_test, model2.predict(x_test),color='g')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()])
plt.show()

# 显示图形
plt.show()

# 计算拟合度
# 使用LinearRegression().score函数计算X和y之间的拟合度并打印拟合度
print(r'使用LinearRegression().score函数计算X和y之间的拟合度并打印拟合度')
score2 = model2.score(x_test, y_test)
print(score2)


# Ridge
from sklearn.linear_model import Ridge,RidgeCV
print('--------------------Ridge a=1--------------------')
model_Ridge = Ridge(alpha=1.0)
model_Ridge.fit(x_train, y_train)   # 线性回归建模

# 训练后模型截距
print('训练后模型截距')
print(model_Ridge.intercept_)

# 训练后模型权重
print('训练后模型权重')
print(model_Ridge.coef_) 

m = model_Ridge.coef_
b = model_Ridge.intercept_
# 根据得到的系数和截距打印回归方程print(' y = {0} * x + {1}'.format(m, b))
print(r'根据得到的系数和截距打印回归方程')
print(' y = {0} * x + {1}'.format(m, b))

# 使用模型预测
plt.scatter(y_test, model_Ridge.predict(x_test),color='g')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()])
plt.show()

# 显示图形
plt.show()

# 计算拟合度
# 使用.score函数计算X和y之间的拟合度并打印拟合度
print(r'使用.score函数计算X和y之间的拟合度并打印拟合度')
score_Ridge = model_Ridge.score(x_test, y_test)
print(score_Ridge)

print('--------------------Ridge a=10--------------------')
model_Ridge = Ridge(alpha=10.0)
model_Ridge.fit(x_train, y_train)   # 线性回归建模

# 训练后模型截距
print('训练后模型截距')
print(model_Ridge.intercept_)

# 训练后模型权重
print('训练后模型权重')
print(model_Ridge.coef_) 

m = model_Ridge.coef_
b = model_Ridge.intercept_
# 根据得到的系数和截距打印回归方程print(' y = {0} * x + {1}'.format(m, b))
print(r'根据得到的系数和截距打印回归方程')
print(' y = {0} * x + {1}'.format(m, b))

# 使用模型预测
plt.scatter(y_test, model_Ridge.predict(x_test),color='g')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()])
plt.show()

# 显示图形
plt.show()

# 计算拟合度
# 使用.score函数计算X和y之间的拟合度并打印拟合度
print(r'使用.score函数计算X和y之间的拟合度并打印拟合度')
score_Ridge = model_Ridge.score(x_test, y_test)
print(score_Ridge)

# Lasso
from sklearn.linear_model import Lasso
print('--------------------Lasso a=1--------------------')
model_Lasso = Lasso(alpha=1.0)
model_Lasso.fit(x_train, y_train)   # 线性回归建模

# 训练后模型截距
print('训练后模型截距')
print(model_Lasso.intercept_)

# 训练后模型权重
print('训练后模型权重')
print(model_Lasso.coef_) 

m = model_Lasso.coef_
b = model_Lasso.intercept_
# 根据得到的系数和截距打印回归方程print(' y = {0} * x + {1}'.format(m, b))
print(r'根据得到的系数和截距打印回归方程')
print(' y = {0} * x + {1}'.format(m, b))

# 使用模型预测
plt.scatter(y_test, model_Lasso.predict(x_test),color='g')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()])
plt.show()

# 显示图形
plt.show()

# 计算拟合度
# 使用.score函数计算X和y之间的拟合度并打印拟合度
print(r'使用.score函数计算X和y之间的拟合度并打印拟合度')
score_Lasso = model_Lasso.score(x_test, y_test)
print(score_Lasso)

print('--------------------Lasso a=10--------------------')
model_Lasso = Lasso(alpha=10.0)
model_Lasso.fit(x_train, y_train)   # 线性回归建模

# 训练后模型截距
print('训练后模型截距')
print(model_Lasso.intercept_)

# 训练后模型权重
print('训练后模型权重')
print(model_Lasso.coef_) 

m = model_Lasso.coef_
b = model_Lasso.intercept_
# 根据得到的系数和截距打印回归方程print(' y = {0} * x + {1}'.format(m, b))
print(r'根据得到的系数和截距打印回归方程')
print(' y = {0} * x + {1}'.format(m, b))

# 使用模型预测
plt.scatter(y_test, model_Lasso.predict(x_test),color='g')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()])
plt.show()

# 显示图形
plt.show()

# 计算拟合度
# 使用.score函数计算X和y之间的拟合度并打印拟合度
print(r'使用.score函数计算X和y之间的拟合度并打印拟合度')
score_Lasso = model_Lasso.score(x_test, y_test)
print(score_Lasso)


# ElasticNet
from sklearn.linear_model import ElasticNet
print('--------------------ElasticNet a=1--------------------')
model_ElasticNet = ElasticNet(alpha=1.0)
model_ElasticNet.fit(x_train, y_train)   # 线性回归建模

# 训练后模型截距
print('训练后模型截距')
print(model_ElasticNet.intercept_)

# 训练后模型权重
print('训练后模型权重')
print(model_ElasticNet.coef_) 

m = model_ElasticNet.coef_
b = model_ElasticNet.intercept_
# 根据得到的系数和截距打印回归方程print(' y = {0} * x + {1}'.format(m, b))
print(r'根据得到的系数和截距打印回归方程')
print(' y = {0} * x + {1}'.format(m, b))

# 使用模型预测
plt.scatter(y_test, model_ElasticNet.predict(x_test),color='g')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()])
plt.show()

# 显示图形
plt.show()

# 计算拟合度
# 使用.score函数计算X和y之间的拟合度并打印拟合度
print(r'使用.score函数计算X和y之间的拟合度并打印拟合度')
score_ElasticNet = model_ElasticNet.score(x_test, y_test)
print(score_ElasticNet)

print('--------------------ElasticNet a=10--------------------')
model_ElasticNet = ElasticNet(alpha=10.0)
model_ElasticNet.fit(x_train, y_train)   # 线性回归建模

# 训练后模型截距
print('训练后模型截距')
print(model_ElasticNet.intercept_)

# 训练后模型权重
print('训练后模型权重')
print(model_ElasticNet.coef_) 

m = model_ElasticNet.coef_
b = model_ElasticNet.intercept_
# 根据得到的系数和截距打印回归方程print(' y = {0} * x + {1}'.format(m, b))
print(r'根据得到的系数和截距打印回归方程')
print(' y = {0} * x + {1}'.format(m, b))

# 使用模型预测
plt.scatter(y_test, model_ElasticNet.predict(x_test),color='g')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()])
plt.show()

# 显示图形
plt.show()

# 计算拟合度
# 使用.score函数计算X和y之间的拟合度并打印拟合度
print(r'使用.score函数计算X和y之间的拟合度并打印拟合度')
score_ElasticNet = model_ElasticNet.score(x_test, y_test)
print(score_ElasticNet)
```
